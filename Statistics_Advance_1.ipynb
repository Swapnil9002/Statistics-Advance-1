#1. What is a random variable in probability theory?
 - A **random variable** is a numerical value assigned to the outcome of a **random experiment**.

* **Discrete**: Takes countable values (e.g., number of heads in coin toss).
* **Continuous**: Takes any value in a range (e.g., height of a person).

It helps us **quantify uncertainty** and is linked to a **probability distribution**.
#2. What are the types of random variables?
 - The two main types of random variables are:

1. **Discrete Random Variable**

   * Takes **countable values** (e.g., 0, 1, 2, ...).
   * Example: Number of heads in 3 coin tosses.

2. **Continuous Random Variable**

   * Takes **any value within a range** (uncountably infinite).
   * Example: Height of a person (like 165.3 cm).

#3.What is the difference between discrete and continuous distributions)
 -  The key differences between **discrete** and **continuous** distributions are:

| Feature                        | Discrete Distribution                     | Continuous Distribution                        |
| ------------------------------ | ----------------------------------------- | ---------------------------------------------- |
| **Type of values**             | Countable (e.g., 0, 1, 2...)              | Any value in a range (e.g., 1.2, 3.56, ...)    |
| **Function used**              | Probability Mass Function (**PMF**)       | Probability Density Function (**PDF**)         |
| **Example**                    | Number of students in a class             | Height of students in a class                  |
| **Probability of exact value** | Has a positive value (e.g., P(X=2) = 0.3) | Always zero (P(X=2) = 0); use area under curve |
| **Graph**                      | Bar graph                                 | Smooth curve                                   |

#4.What are probability distribution functions (PDF)?
 - A **Probability Distribution Function (PDF)** shows how probabilities are spread across values of a random variable.

* For **discrete** variables: called **PMF** (gives exact probabilities).
* For **continuous** variables: called **PDF** (gives probability over an interval, area under curve).

In short:

* **PMF** → exact values (e.g., P(X = 2))
* **PDF** → range of values (e.g., P(1 < X < 3))

#5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
 - ### Difference between **CDF** and **PDF**:

| Feature       | **PDF/PMF**                               | **CDF**                              |
| ------------- | ----------------------------------------- | ------------------------------------ |
| **Full Form** | Probability Density/Mass Function         | Cumulative Distribution Function     |
| **Gives**     | Probability at a specific value (PDF/PMF) | Probability up to a value (P(X ≤ x)) |
| **Type**      | PDF → continuous, PMF → discrete          | Applies to both types                |
| **Values**    | Can vary widely                           | Always increases from 0 to 1         |
| **Example**   | PDF: P(X = 2) or f(x)                     | CDF: P(X ≤ 2)                        |

### In short:

* **PDF/PMF** → probability **at** a point
* **CDF** → probability **up to** a point

#6.What is a discrete uniform distribution?
 - A **discrete uniform distribution** is a probability distribution where **all outcomes are equally likely**.

### Key Features:

* Each value has the **same probability**.
* Used when there are **finite, countable** outcomes.

### Example:

Rolling a fair six-sided die:

* Outcomes: 1, 2, 3, 4, 5, 6
* Each has probability $\frac{1}{6}$

### Formula:

If there are $n$ possible outcomes:

$$
P(X = x) = \frac{1}{n}
$$

### In short:

A **discrete uniform distribution** gives **equal chance** to all possible values.
#7.What are the key properties of a Bernoulli distribution?
 - The **Bernoulli distribution** is a discrete distribution for a random experiment with only **two outcomes**: **success (1)** or **failure (0)**.

### Key Properties:

1. **Two outcomes**: 1 (success) or 0 (failure)
2. **One trial**: Single experiment (e.g., coin toss)
3. **Probability**:

   * $P(X = 1) = p$
   * $P(X = 0) = 1 - p$, where $0 \leq p \leq 1$
4. **Mean**: $\mu = p$
5. **Variance**: $\sigma^2 = p(1 - p)$

### Example:

Tossing a coin:

* Head (1) with $p = 0.5$
* Tail (0) with $1 - p = 0.5$

### In short:

The **Bernoulli distribution** models a **single yes/no** or **success/failure** event.
#8.What is the binomial distribution, and how is it used in probability?
 - The **binomial distribution** models the number of **successes** in a fixed number of **independent** trials, each with the **same probability** of success.

---

### **Key Features:**

* **n** = number of trials
* **p** = probability of success in each trial
* **X** = number of successes

---

### **Formula:**

$$
P(X = k) = \binom{n}{k} p^k (1 - p)^{n - k}
$$

Where:

* $\binom{n}{k}$ = number of ways to choose $k$ successes from $n$ trials
* $k$ = 0, 1, 2, ..., n

---

### **Example:**

Flipping a coin 5 times, what's the probability of getting exactly 2 heads?
Here: $n = 5$, $p = 0.5$, $k = 2$

---

### **Used in:**

* Quality control (defective items)
* Medical trials (number of patients cured)
* Surveys (yes/no responses)

---

### **In short:**

The **binomial distribution** gives the probability of getting **k successes in n trials**, with **constant success probability** and **independent trials**.

#9. What is the Poisson distribution and where is it applied?
 - The **Poisson distribution** models the number of times an event occurs in a **fixed interval** of time or space, when the events happen **independently** and at a **constant average rate**.

---

### **Key Features:**

* Used for **counting events** (e.g., arrivals, accidents, emails).
* Defined by **λ (lambda)** = average rate of occurrence.
* Events occur **randomly and independently**.

---

### **Formula:**

$$
P(X = k) = \frac{e^{-λ} λ^k}{k!}
$$

Where:

* $X$ = number of occurrences
* $λ$ = mean number of occurrences
* $e \approx 2.718$
* $k$ = 0, 1, 2, ...

---

### **Applications:**

* Number of customer arrivals per hour
* Calls at a call center
* Accidents at a traffic junction
* Emails received per minute

---

### **In short:**

The **Poisson distribution** is used to model the **number of random events** in a fixed time or space, given a **known average rate**.
#10. What is a continuous uniform distribution?
 - A **continuous uniform distribution** is a probability distribution where **all values within a given interval are equally likely**.

---

### **Key Features:**

* Defined over an interval $[a, b]$
* Every value between $a$ and $b$ has the **same probability density**
* Shape: a **flat** horizontal line

---

### **PDF Formula:**

$$
f(x) = \frac{1}{b - a} \quad \text{for } a \le x \le b
$$

---

### **Mean:**

$$
\mu = \frac{a + b}{2}
$$

### **Variance:**

$$
\sigma^2 = \frac{(b - a)^2}{12}
$$

---

### **Example:**

If a bus arrives randomly between 10:00 and 10:30, the waiting time follows a **uniform distribution** from 0 to 30 minutes.

---

### **In short:**

A **continuous uniform distribution** gives **equal chance** to all values in a continuous range $[a, b]$.
#11.What are the characteristics of a normal distribution?
 - ### **Characteristics of a Normal Distribution:**

1. **Bell-shaped curve**

   * Symmetrical around the mean.

2. **Mean = Median = Mode**

   * All located at the center.

3. **Defined by two parameters**

   * **Mean (μ)**: center of the distribution
   * **Standard deviation (σ)**: spread or width

4. **68-95-99.7 Rule (Empirical Rule)**

   * 68% of data within ±1σ
   * 95% within ±2σ
   * 99.7% within ±3σ

5. **Total area under the curve = 1**

   * Represents total probability.

6. **Tails extend infinitely**

   * But never touch the horizontal axis.

---

### **In short:**

A **normal distribution** is a **symmetric, bell-shaped** curve where most values cluster around the mean, and probabilities decrease as you move away from the center.
#12.What is the standard normal distribution, and why is it important?
 - ### **Standard Normal Distribution:**

The **standard normal distribution** is a **normal distribution** with:

* **Mean (μ) = 0**
* **Standard deviation (σ) = 1**

---

### **Why It's Important:**

1. **Simplifies calculations**

   * Probabilities and z-scores are easily found using standard normal tables.

2. **Used for comparison**

   * Any normal distribution can be **converted** to standard normal using:

     $$
     Z = \frac{X - \mu}{\sigma}
     $$

3. **Foundation for statistical tests**

   * Widely used in **hypothesis testing** and **confidence intervals**.

---

### **In short:**

The **standard normal distribution** is a special case of the normal distribution (mean 0, standard deviation 1) and is essential for **solving probability problems** and **standardizing data**.
#13.  What is the Central Limit Theorem (CLT), and why is it critical in statistics?
 - ### **Central Limit Theorem (CLT):**

The **Central Limit Theorem** states that:

> When you take a **large enough sample** from any population (with a finite mean and variance), the **sampling distribution** of the **sample mean** will be **approximately normal**, regardless of the population's original distribution.

---

### **Why It's Critical:**

1. ✅ **Makes normal distribution widely usable**

   * Even if data isn't normal, **sample means** will follow a normal pattern.

2. ✅ **Enables statistical inference**

   * Helps construct **confidence intervals** and perform **hypothesis testing**.

3. ✅ **Applies to real-world data**

   * Works for most types of data when sample size is **large (n ≥ 30)**.

---

### **In short:**

The **CLT** allows us to use **normal distribution tools** for analyzing averages, even when the data isn't normally distributed — making it a **foundation of modern statistics**.

#14.How does the Central Limit Theorem relate to the normal distribution?
 - ### **Relation Between CLT and Normal Distribution:**

The **Central Limit Theorem (CLT)** explains **why the normal distribution appears so often** in statistics.

---

### 🔁 **Key Connection:**

CLT says that:

> As the **sample size increases**, the **distribution of the sample mean** approaches a **normal distribution**, even if the population itself is **not normal**.

---

### 🔍 So, CLT helps us:

* Use **normal distribution** to make inferences about **sample means**
* Apply statistical tools like **z-scores, confidence intervals**, and **hypothesis tests**

---

### **In short:**

The **CLT links real-world data to the normal distribution**, making it possible to apply **normal-based methods** even when the original data isn’t normally distributed — as long as the sample size is large.

#15.What is the application of Z statistics in hypothesis testing?
 - ### **Application of Z-Statistic in Hypothesis Testing:**

The **Z-statistic** is used in hypothesis testing to determine how far a sample mean is from the population mean **in terms of standard deviations**.

---

### ✅ **When to Use Z-Test:**

* Population standard deviation (**σ**) is known
* Sample size is **large** (usually $n \geq 30$)

---

### **Steps in Hypothesis Testing using Z-Statistic:**

1. **State hypotheses**

   * Null ($H_0$): no effect or difference
   * Alternate ($H_1$): there is an effect or difference

2. **Calculate Z-statistic**

   $$
   Z = \frac{\bar{X} - \mu}{\sigma / \sqrt{n}}
   $$

   * $\bar{X}$ = sample mean
   * $\mu$ = population mean
   * $\sigma$ = population standard deviation
   * $n$ = sample size

3. **Compare with critical Z-value**

   * Based on significance level (e.g., 1.96 for 5% level in two-tailed test)

4. **Make decision**

   * If $|Z|$ > critical value → **Reject $H_0$**
   * Else → **Fail to reject $H_0$**

---

### **In short:**

The **Z-statistic** helps test if a sample mean is significantly different from a known population mean, especially when the population standard deviation is known and the sample size is large.
#16.  How do you calculate a Z-score, and what does it represent?
 - ### **Z-Score: Definition**

A **Z-score** tells you **how many standard deviations** a data point is from the **mean**.

---

### **Formula:**

$$
Z = \frac{X - \mu}{\sigma}
$$

Where:

* $X$ = individual value
* $\mu$ = population mean
* $\sigma$ = population standard deviation

---

### **What It Represents:**

* **Z = 0** → value is at the mean
* **Z > 0** → value is above the mean
* **Z < 0** → value is below the mean

---

### **Example:**

If a test score $X = 85$, the class average $\mu = 75$, and standard deviation $\sigma = 5$:

$$
Z = \frac{85 - 75}{5} = 2
$$

✅ This means the score is **2 standard deviations above** the mean.

---

### **In short:**

The **Z-score** standardizes data, showing how far a value is from the mean in terms of **standard deviations**.
#17.What are point estimates and interval estimates in statistics?
 - ### **Point Estimate vs. Interval Estimate**

#### ✅ **Point Estimate:**

* A **single value** used to **estimate a population parameter**.
* Example: Sample mean ($\bar{x}$) estimates population mean ($\mu$).

> 🔹 **Quick but uncertain**

---

#### ✅ **Interval Estimate:**

* A **range of values** that likely contains the population parameter.
* Usually expressed as a **confidence interval**.

$$
\text{Confidence Interval} = \bar{x} \pm \text{Margin of Error}
$$

> 🔹 **More informative and reliable**

---

### **In short:**

* **Point estimate** = one best guess
* **Interval estimate** = range with a confidence level (e.g., 95%)
#18. What is the significance of confidence intervals in statistical analysis?
 - ### **Significance of Confidence Intervals in Statistics:**

A **confidence interval (CI)** gives a **range of values** within which a **population parameter** is likely to lie, based on sample data.

---

### ✅ **Why Confidence Intervals Are Important:**

1. **Shows estimation uncertainty**

   * Unlike a single point estimate, a CI tells **how precise** the estimate is.

2. **Provides a confidence level**

   * Example: A **95% CI** means we are 95% confident the true value lies within the interval.

3. **Supports better decision-making**

   * Helps assess **reliability** of results in research and business.

4. **Used in hypothesis testing**

   * If the CI **excludes** a hypothesized value, we may **reject** the null hypothesis.

---

### **In short:**

Confidence intervals give a **reliable range** (not just a guess) for estimating population values and help assess how **trustworthy** the sample results are.
#19. What is the relationship between a Z-score and a confidence interval?
 - ### ✅ **Relationship Between Z-Score and Confidence Interval:**

A **Z-score** is used to calculate a **confidence interval** when the population standard deviation is known and the sample size is large.

---

### 📌 **Confidence Interval Formula using Z-score:**

$$
\text{CI} = \bar{x} \pm Z \cdot \frac{\sigma}{\sqrt{n}}
$$

Where:

* $\bar{x}$ = sample mean
* $Z$ = Z-score corresponding to the confidence level
* $\sigma$ = population standard deviation
* $n$ = sample size

---

### 🔢 **Common Z-scores for Confidence Levels:**

| Confidence Level | Z-score |
| ---------------- | ------- |
| 90%              | 1.645   |
| 95%              | 1.96    |
| 99%              | 2.576   |

---

### ✅ **In short:**

The **Z-score** determines how wide the **confidence interval** is — higher confidence levels require **larger Z-scores**, resulting in **wider intervals**.
#20. How are Z-scores used to compare different distributions?
 - ### ✅ **Z-Scores for Comparing Different Distributions:**

**Z-scores** allow you to **standardize** values from different distributions so they can be **compared directly**, even if the original scales are different.

---

### 📌 **How It Works:**

$$
Z = \frac{X - \mu}{\sigma}
$$

Where:

* $X$ = individual value
* $\mu$ = mean of the distribution
* $\sigma$ = standard deviation

---

### 🧠 **Why It's Useful:**

* Converts values to a **common scale** (standard normal distribution).
* Shows how **far** and in **what direction** a value is from its mean.

---

### 📊 **Example:**

* Student A scores 85 in Math (mean = 75, SD = 5) → $Z = \frac{85 - 75}{5} = 2$
* Student B scores 90 in English (mean = 80, SD = 10) → $Z = \frac{90 - 80}{10} = 1$

✅ Student A performed **better relative to their group**, despite a lower raw score.

---

### **In short:**

**Z-scores** standardize different distributions, making it easy to **compare performances or values** across different scales or datasets.
#21. What are the assumptions for applying the Central Limit Theorem?
 - ### ✅ **Assumptions for Applying the Central Limit Theorem (CLT):**

1. **Independent Observations**

   * Each sample value must be collected independently.

2. **Identically Distributed**

   * All observations come from the same population distribution.

3. **Sample Size is Large**

   * Typically, $n \geq 30$ is considered sufficient.
   * Smaller samples may work **if the population is already normal**.

4. **Finite Mean and Variance**

   * The population must have a defined (finite) mean (μ) and variance (σ²).

---

### 🧠 **Why These Matter:**

* Violating these assumptions may lead to incorrect conclusions, especially with small sample sizes or skewed data.

---

### **In short:**

To apply the CLT, use a **large, independent, and identically distributed sample** from a population with **finite mean and variance**.
#22.What is the concept of expected value in a probability distribution?
 - ### ✅ **Expected Value in a Probability Distribution:**

The **expected value (EV)** is the **average outcome** you would expect if a random experiment is repeated **many times**. It represents the **long-term mean** of a random variable.

---

### 📌 **Formula:**

* **For Discrete Random Variables**:

  $$
  E(X) = \sum [x \cdot P(x)]
  $$

* **For Continuous Random Variables**:

  $$
  E(X) = \int x \cdot f(x) \, dx
  $$

Where:

* $x$ = possible outcome
* $P(x)$ or $f(x)$ = probability or probability density function

---

### 📊 **Example (Discrete):**

If you roll a fair 6-sided die:

$$
E(X) = 1 \cdot \frac{1}{6} + 2 \cdot \frac{1}{6} + \dots + 6 \cdot \frac{1}{6} = 3.5
$$

---

### **In short:**

The **expected value** is the **weighted average** of all possible outcomes, giving a measure of the **center** of a probability distribution.

#23. How does a probability distribution relate to the expected outcome of a random variable?
 - ### ✅ **Relationship Between Probability Distribution and Expected Outcome:**

A **probability distribution** shows **all possible values** a random variable can take and their **associated probabilities**.

The **expected outcome** (or **expected value**) is the **weighted average** of those values, using the probabilities from the distribution.

---

### 📌 **In simple terms:**

* The **probability distribution** tells you **what could happen** and **how likely**.
* The **expected value** tells you what you’d get **on average** if you repeated the experiment many times.

---

### 🧠 **Example:**

For a fair 6-sided die:

* Distribution: Each outcome (1–6) has $P(x) = \frac{1}{6}$
* Expected value:

  $$
  E(X) = \sum x \cdot P(x) = 3.5
  $$

---

### **In short:**

The **expected value** is calculated **from** the **probability distribution** and gives the **average result** of a random variable over time.

```python
#3.Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.
def bernoulli_pdf(x, p):
    """
    Calculate the PDF of a Bernoulli distribution.

    Parameters:
    x (int): The outcome (0 or 1)
    p (float): Probability of success (0 ≤ p ≤ 1)

    Returns:
    float: Probability of getting outcome x
    """
    if x not in [0, 1]:
        return 0.0
    return p if x == 1 else 1 - p

# Example usage:
print("P(X=1):", bernoulli_pdf(1, 0.7))  # Success
print("P(X=0):", bernoulli_pdf(0, 0.7))  # Failure
```

```python
#6.Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define range for the discrete uniform distribution (e.g., dice roll 1 to 6)
low = 1
high = 7  # upper bound is exclusive in scipy

# Create the distribution object
dist = randint(low, high)

# Values where CDF is evaluated
x = np.arange(low - 1, high + 1)

# Compute CDF values
cdf = dist.cdf(x)

# Plot the CDF
plt.step(x, cdf, where='post', label='CDF')
plt.title("CDF of Discrete Uniform Distribution (1–6)")
plt.xlabel("Value")
plt.ylabel("Cumulative Probability")
plt.xticks(x)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

```python
#9. Write a Python function to calculate Z-scores from a dataset and plot them
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_zscores(data):
    # Calculate mean and standard deviation
    mean = np.mean(data)
    std = np.std(data)

    # Calculate Z-scores
    z_scores = (data - mean) / std

    # Plot original data and Z-scores
    plt.figure(figsize=(10, 5))

    # Plot original data
    plt.subplot(1, 2, 1)
    plt.hist(data, bins=20, edgecolor='black', alpha=0.7)
    plt.title("Original Data")
    plt.xlabel("Value")
    plt.ylabel("Frequency")
    plt.grid(True, linestyle='--', alpha=0.5)

    # Plot Z-scores
    plt.subplot(1, 2, 2)
    plt.hist(z_scores, bins=20, edgecolor='black', alpha=0.7, color='orange')
    plt.title("Z-Scores")
    plt.xlabel("Z-score")
    plt.ylabel("Frequency")
    plt.grid(True, linestyle='--', alpha=0.5)

    plt.tight_layout()
    plt.show()

    return z_scores

# Example usage
data = np.random.normal(50, 10, 1000)  # Generate sample data
z_scores = calculate_and_plot_zscores(data)
```

```python
#10.  Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.
import numpy as np
import matplotlib.pyplot as plt

# Parameters
sample_size = 30         # Size of each sample (n)
num_samples = 1000       # Number of samples
scale = 2.0              # Exponential distribution parameter

# Store means of samples
sample_means = []

# Generate sample means from exponential distribution
for _ in range(num_samples):
    sample = np.random.exponential(scale=scale, size=sample_size)
    sample_means.append(np.mean(sample))

# Plot histogram of sample means
plt.hist(sample_means, bins=30, density=True, edgecolor='black', alpha=0.7)
plt.title(f'CLT Demo: Sampling Means from Exponential (n={sample_size})')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()

```

```python
#11.Simulate multiple samples from a normal distribution and verify the Central Limit Theorem
import numpy as np
import matplotlib.pyplot as plt

# Parameters
population_mean = 100
population_std = 15
sample_size = 30
num_samples = 1000

# Generate multiple samples and compute their means
sample_means = []

for _ in range(num_samples):
    sample = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)
    sample_mean = np.mean(sample)
    sample_means.append(sample_mean)

# Plot the histogram of sample means
plt.hist(sample_means, bins=30, density=True, edgecolor='black', alpha=0.7)
plt.title(f'Central Limit Theorem Verification\nSample Size = {sample_size}, Samples = {num_samples}')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
#12.Write a Python function to calculate and plot the standard normal distribution (mean = 0, std = 1)
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_standard_normal():
    # Generate x values from -4 to 4
    x = np.linspace(-4, 4, 500)

    # Calculate the PDF of the standard normal distribution
    y = norm.pdf(x, loc=0, scale=1)

    # Plot
    plt.plot(x, y, label='Standard Normal PDF', color='blue')
    plt.title('Standard Normal Distribution (μ = 0, σ = 1)')
    plt.xlabel('Z-score')
    plt.ylabel('Probability Density')
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.legend()
    plt.show()

# Call the function
plot_standard_normal()
```

```python
#13.Generate random variables and calculate their corresponding probabilities using the binomial distribution5
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters for the Binomial Distribution
n = 10      # number of trials
p = 0.5     # probability of success
size = 1000 # number of random variables to generate

# Generate random variables from Binomial distribution
data = np.random.binomial(n, p, size)

# Unique values and their frequencies
unique, counts = np.unique(data, return_counts=True)

# Calculate PMF (probability mass function) for each unique value
pmf_values = binom.pmf(unique, n, p)

# Display results
print("Value\tObserved Count\tTheoretical PMF")
for val, count, pmf in zip(unique, counts, pmf_values):
    print(f"{val}\t{count}\t\t{pmf:.4f}")

# Plot histogram and PMF
plt.bar(unique, counts / size, width=0.6, alpha=0.6, label='Observed (Relative Freq)', edgecolor='black')
plt.plot(unique, pmf_values, 'o-', color='red', label='Theoretical PMF')
plt.title(f'Binomial Distribution (n={n}, p={p})')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
#14.Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(x, mean, std):
    """Calculate the Z-score of a given value."""
    return (x - mean) / std

# Example data
data_point = 85
mean = 75
std_dev = 5

# Calculate Z-score
z = calculate_z_score(data_point, mean, std_dev)
print(f"Z-score for x = {data_point}: {z:.2f}")

# Plot standard normal distribution
x_values = np.linspace(-4, 4, 500)
y_values = norm.pdf(x_values)

plt.plot(x_values, y_values, label='Standard Normal PDF', color='blue')
plt.axvline(z, color='red', linestyle='--', label=f'Z = {z:.2f}')
plt.title('Standard Normal Distribution and Z-score')
plt.xlabel('Z-score')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
#15. Implement hypothesis testing using Z-statistics for a sample dataset.
import numpy as np
from scipy.stats import norm

# Sample data
sample = [82, 85, 88, 90, 87, 84, 83, 86, 89, 85]
sample_mean = np.mean(sample)
n = len(sample)

# Known population parameters
mu_0 = 80     # population mean under H0
sigma = 5     # known population standard deviation
alpha = 0.05  # significance level

# Step 1: Calculate Z-statistic
z = (sample_mean - mu_0) / (sigma / np.sqrt(n))

# Step 2: Compute p-value for two-tailed test
p_value = 2 * (1 - norm.cdf(abs(z)))

# Step 3: Decision
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-statistic: {z:.2f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("Reject the null hypothesis (H₀)")
else:
    print("Fail to reject the null hypothesis (H₀)")
```

```python
#16.Create a confidence interval for a dataset using Python and interpret the result
import numpy as np
from scipy import stats

# Sample dataset
data = [12, 15, 14, 10, 13, 16, 14, 15, 13, 14]
n = len(data)
mean = np.mean(data)
std_err = stats.sem(data)  # Standard error of the mean

# 95% Confidence Interval using t-distribution
confidence = 0.95
df = n - 1  # degrees of freedom
t_critical = stats.t.ppf((1 + confidence) / 2, df)

margin_of_error = t_critical * std_err
lower_bound = mean - margin_of_error
upper_bound = mean + margin_of_error

print(f"Sample Mean: {mean:.2f}")
print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")

```

```python
#17.Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Step 1: Generate data from a normal distribution
np.random.seed(42)  # for reproducibility
mu, sigma = 50, 10  # population mean and std dev
sample_size = 100
data = np.random.normal(mu, sigma, sample_size)

# Step 2: Calculate sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # unbiased estimator
confidence_level = 0.95
alpha = 1 - confidence_level

# Step 3: Calculate the margin of error
z_score = stats.norm.ppf(1 - alpha/2)  # for 95% CI
margin_of_error = z_score * (sample_std / np.sqrt(sample_size))

# Step 4: Confidence interval
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

# Step 5: Print results
print(f"Sample Mean = {sample_mean:.2f}")
print(f"95% Confidence Interval = ({ci_lower:.2f}, {ci_upper:.2f})")

# Step 6: Plot histogram and CI
plt.hist(data, bins=20, edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='red', linestyle='--', label='Sample Mean')
plt.axvline(ci_lower, color='green', linestyle='--', label='95% CI Lower')
plt.axvline(ci_upper, color='green', linestyle='--', label='95% CI Upper')
plt.title("Histogram of Normally Distributed Data")
plt.legend()
plt.show()
```

```python
#23. Implement Z-score normalization on a dataset and visualize its transformation.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore

# Step 1: Generate sample data
np.random.seed(42)
data = np.random.normal(loc=70, scale=10, size=1000)  # Mean 70, std 10

# Step 2: Z-score normalization
z_normalized = zscore(data)

# Step 3: Print summary
print(f"Original Data → Mean: {np.mean(data):.2f}, Std Dev: {np.std(data):.2f}")
print(f"Z-Score Normalized Data → Mean: {np.mean(z_normalized):.2f}, Std Dev: {np.std(z_normalized):.2f}")

# Step 4: Visualization
plt.figure(figsize=(12, 5))

# Histogram of Original Data
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title("Original Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.axvline(np.mean(data), color='red', linestyle='--', label='Mean')
plt.legend()

# Histogram of Z-Normalized Data
plt.subplot(1, 2, 2)
plt.hist(z_normalized, bins=30, color='lightgreen', edgecolor='black')
plt.title("Z-Score Normalized Data")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.axvline(0, color='red', linestyle='--', label='Mean = 0')
plt.legend()

plt.tight_layout()
plt.show()

```

```python
#24.Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution
import numpy as np
from scipy.stats import skew, kurtosis

def analyze_normal_distribution_skew_kurtosis(mean=0, std_dev=1, size=1000):
    # Step 1: Generate data
    data = np.random.normal(loc=mean, scale=std_dev, size=size)

    # Step 2: Calculate skewness and kurtosis
    skewness = skew(data)
    kurt = kurtosis(data, fisher=True)  # Fisher=True: subtract 3 so normal distribution ≈ 0

    # Step 3: Print results
    print(f"Skewness: {skewness:.4f}")
    print(f"Kurtosis: {kurt:.4f} (Fisher’s definition, Normal ≈ 0)")

    return skewness, kurt

# Example usage
analyze_normal_distribution_skew_kurtosis(mean=50, std_dev=10, size=1000)
```
